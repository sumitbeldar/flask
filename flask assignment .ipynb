In Flask, App routing refers to the process of mapping URLs to specific functions or view handlers in your Flask application. This is done using decorators, which are used to define URL patterns and associate them with specific functions.

For example, let's say you want to create a web application with two pages, one for displaying the homepage and another for displaying a contact form. In Flask, you would define these pages as functions and use the @app.route() decorator to associate each function with a specific URL.

In Flask, the url_for() function is used for URL building. It generates a URL for a given endpoint with the specified method.
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def home():
    return 'This is the home page.'

@app.route('/contact')
def contact():
    return 'This is the contact page.'

if __name__ == '__main__':
    with app.test_request_context():
        print(url_for('home'))
        print(url_for('contact'))
